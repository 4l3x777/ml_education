# 1.3

Рассмотрим матрицу

$$
A = \begin{bmatrix}
1 & 2 & 3 \\
3 & 6 & p \\
5 & 10 & q
\end{bmatrix}
$$

## **Задача**

Определить, при каких значениях параметров \( p \) и \( q \) **ранг матрицы \( A \) равен единице**.

---

## **Решение**

**Ранг** матрицы равен 1, если все её строки линейно зависимы (пропорциональны друг другу).

### **Проверка пропорциональности строк**

Пусть первая строка — эталонная:

\[
(1, 2, 3)
\]

**Вторая строка:**

\[
(3, 6, p)
\]

Должна быть пропорциональна первой строке, поэтому

\[
$$
3 = 3 \cdot 1,\qquad 6 = 3 \cdot 2,\qquad p = 3 \cdot 3 = 9
$$
\]

**Третья строка:**

\[
(5, 10, q)
\]

Также должна быть пропорциональна первой строке, поэтому

\[
$$
5 = 5 \cdot 1,\qquad 10 = 5 \cdot 2,\qquad q = 5 \cdot 3 = 15
$$
\]

---

## **Ответ**

**Ранг матрицы \( A \) равен 1 только при**

\[
$$
\boxed{p = 9, ~ q = 15}
$$
\]

---

## **Проверка свойств**

- **При этих значениях** все строки пропорциональны друг другу, и ранг действительно равен 1.
- **Если хотя бы одно из равенств нарушено** (\($$ p \neq 9 $$\) или \($$ q \neq 15 $$\)), матрица будет иметь ранг 2 или 3.

# Переход между базисами в ℝ²

## **Дано**

- **Векторы:**  
  \($$ a_1 = \begin{bmatrix} 2 \\ -5 \end{bmatrix}, \quad a_2 = \begin{bmatrix} -1 \\ 3 \end{bmatrix} $$\)
- **Вектор x:**  
  \($$ x = \begin{bmatrix} 1 \\ -4 \end{bmatrix} $$\)
- **Новый базис:**  
  \($$ B = \{a_1, a_2\} $$\)
- **Координаты вектора y в базисе B:** \($$ [y]_B = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $$\)

---

## **(a) Координаты вектора x в базисе B**

### **Проверка линейной независимости базиса B**

\[
$$
\det \begin{pmatrix} 2 & -1 \\ -5 & 3 \end{pmatrix} = 2 \cdot 3 - (-1) \cdot (-5) = 6 - 5 = 1 \neq 0
$$
\]
**Векторы линейно независимы** — образуют базис.

### **Разложение вектора x по базису B**

Пусть \($$ x = \alpha a_1 + \beta a_2 $$\):

\[
\begin{cases}  
2\alpha - \beta = 1 \\  
-5\alpha + 3\beta = -4  
\end{cases}
\]

**Решаем систему:**  
Из первого уравнения выразим \($$\beta = 2\alpha - 1$$\).  
Подставим во второе уравнение:

\[
$$
-5\alpha + 3(2\alpha - 1) = -4 \\  
-5\alpha + 6\alpha - 3 = -4 \\  
\alpha - 3 = -4 \\  
\alpha = -1
$$
\]

Тогда \($$\beta = 2 \cdot (-1) - 1 = -3$$\).

**Ответ:**

\[
$$
\boxed{[x]_B = \begin{bmatrix} -1 \\ -3 \end{bmatrix}}
$$
\]

---

## **(b) Координаты вектора y в стандартном базисе**

**Если \($$ [y]_B = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $$\), то**

\[
$$
y = 1 \cdot a_1 + 1 \cdot a_2 = \begin{bmatrix} 2 \\ -5 \end{bmatrix} + \begin{bmatrix} -1 \\ 3 \end{bmatrix} = \begin{bmatrix} 1 \\ -2 \end{bmatrix}
$$
\]

**Или через матрицу перехода:**

\[
$$
A = \begin{bmatrix} 2 & -1 \\ -5 & 3 \end{bmatrix}
$$
\]
\[
$$
y = A \cdot [y]_B = \begin{bmatrix} 2 & -1 \\ -5 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 1 \\ -2 \end{bmatrix}
$$
\]

**Ответ:**

\[
$$
\boxed{y = \begin{bmatrix} 1 \\ -2 \end{bmatrix}}
$$
\]

---

## **Итог**

- **Векторы a₁, a₂ образуют базис B.**
- **Координаты x в B: \($$ [x]_B = \begin{bmatrix} -1 \\ -3 \end{bmatrix} $$\).**
- **Координаты y в стандартном базисе: \($$ y = \begin{bmatrix} 1 \\ -2 \end{bmatrix} $$\).**

```python
# Больше разборчивый исходник с комментариями для Jupyter Notebook

import numpy as np
import matplotlib.pyplot as plt

# ------------------
# 1. Генерация случайной матрицы
# ------------------

n = 100  # Размер матрицы
A = np.random.randn(n, n)  # Матрица с нормально распределенными элементами

# ------------------
# 2. Вычисление сингулярного разложения (SVD)
# ------------------

# SVD в компактной форме (без избыточных столбцов/строк, если матрица не квадратная)
U, s, Vh = np.linalg.svd(A, full_matrices=False)
# s — массив сингулярных чисел, диагональ матрицы Σ
# Восстанавливаем диагональную матрицу Σ
Sigma = np.diag(s)

# ------------------
# 3. Перебираем ранги аппроксимации от 2 до n
# ------------------

# Создадим список ошибок
errors = []
# Создадим список рангов
ranks = range(2, n + 1)

for r in ranks:
    # Берём первые r сингулярных векторов и значений
    U_r = U[:, :r]
    Sigma_r = np.diag(s[:r])
    Vh_r = Vh[:r, :]
    # Восстанавливаем матрицу малого ранга
    A_r = U_r @ Sigma_r @ Vh_r
    # Ошибка аппроксимации — норма Фробениуса разности
    error = np.linalg.norm(A - A_r, 'fro')
    errors.append(error)

# ------------------
# 4. Визуализация
# ------------------

plt.figure(figsize=(10, 6))  # Размер графика в дюймах
plt.plot(ranks, errors, '-o', markersize=2, color='blue')
plt.xlabel('Ранг аппроксимации, $r$', fontsize=12)
plt.ylabel('Ошибка аппроксимации (норма Фробениуса)', fontsize=12)
plt.title('Зависимость ошибки аппроксимации матрицы от ранга $r$', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()  # Чтобы подписи не выходили за границы
plt.show()
```

Менять значение r при аппроксимации матрицы — это искать компромисс между степенью сжатия (размером данных) и качеством восстановления исходной информации. Чем больше r, тем точнее аппроксимация, но тем больше требуется памяти и вычислительных ресурсов.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float

# Загружаем изображение (например, 512x512x3)
img = img_as_float(io.imread('cat.jpg'))

def svd_compress_channel(channel, k):
    U, s, Vh = np.linalg.svd(channel, full_matrices=False)
    S = np.diag(s[:k])
    return U[:, :k] @ S @ Vh[:k, :]

# Параметры
ks = [5, 10, 20, 40, 80, 100]

fig, axs = plt.subplots(1, len(ks) + 1, figsize=(15,5))

# Исходное изображение
axs[0].imshow(img)
axs[0].set_title('Original')
axs[0].axis('off')

# Сжатие с разными k
for i, k in enumerate(ks, 1):
    compressed = np.zeros_like(img)
    for c in range(3):
        compressed[:, :, c] = svd_compress_channel(img[:, :, c], k)
    compressed = np.clip(compressed, 0, 1)
    axs[i].imshow(compressed)
    axs[i].set_title(f'Rank {k}')
    axs[i].axis('off')

plt.show()
```

\]
\[
$$

$$
\]

**Ответ:**
Критические точки:
$$
\boxed{
\begin{aligned}

# 2.2

Дана функция:  
$$
f(x_1, x_2) = \ln(\sqrt{x_1} + \sqrt{x_2})
$$

Необходимо проверить, что  
$$
x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{1}{2}
$$

---

## 1. Вычисление частных производных

### Производная по \(x_1\):
$$
\frac{\partial f}{\partial x_1} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_1}} = \frac{1}{2\sqrt{x_1} (\sqrt{x_1} + \sqrt{x_2})}
$$

### Производная по \(x_2\):
$$
\frac{\partial f}{\partial x_2} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_2}} = \frac{1}{2\sqrt{x_2} (\sqrt{x_1} + \sqrt{x_2})}
$$

---

## 2. Подставляем в выражение

$$
x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = x_1 \cdot \frac{1}{2\sqrt{x_1} (\sqrt{x_1} + \sqrt{x_2})} + x_2 \cdot \frac{1}{2\sqrt{x_2} (\sqrt{x_1} + \sqrt{x_2})}
$$

$$
= \frac{x_1}{2\sqrt{x_1} (\sqrt{x_1} + \sqrt{x_2})} + \frac{x_2}{2\sqrt{x_2} (\sqrt{x_1} + \sqrt{x_2})}
$$

(Замечаем, что \($$x_1 = (\sqrt{x_1})^2$$\), \($$x_2 = (\sqrt{x_2})^2$$\))

$$
= \frac{(\sqrt{x_1})^2}{2\sqrt{x_1} (\sqrt{x_1} + \sqrt{x_2})} + \frac{(\sqrt{x_2})^2}{2\sqrt{x_2} (\sqrt{x_1} + \sqrt{x_2})}
$$

$$
= \frac{\sqrt{x_1}}{2(\sqrt{x_1} + \sqrt{x_2})} + \frac{\sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})}
$$

$$
= \frac{\sqrt{x_1} + \sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2}
$$

---

## 3. Ответ

$$
\boxed{
x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{1}{2}
}
$$

  $$
  \]

  \[
  $$
  \frac{\partial f_2}{\partial x} = y z, \quad
  \frac{\partial f_2}{\partial y} = x z, \quad
  \frac{\partial f_2}{\partial z} = x y
  $$
  \]

Подставляем в матрицу:

$$
\boxed{
J_f(x, y, z) =

\]

# 2.4

Для функции
$$
f(x) = \frac{1}{3} \|x\|_2^3, \quad x \in \mathbb{R}^n \setminus \{0\}
$$
где \(
$$
\|x\|_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
$$
\) — евклидова норма.

---

## Градиент

Градиент функции \(f\) — это вектор частных производных:

\[
$$
\nabla f(x) = \left( \frac{\partial f}{\partial x_1}, \dots, \frac{\partial f}{\partial x_n} \right)
$$
\]

Вычислим частные производные:

\[
$$
\frac{\partial f}{\partial x_i} = \frac{1}{3} \cdot \frac{3}{2} \left( \sum_{k=1}^n x_k^2 \right)^{1/2} \cdot 2x_i = \|x\|_2\, x_i
$$
\]

Таким образом,

\[
$$
\boxed{
\nabla f(x) = \|x\|_2\, x
}
$$
\]

---

## Первый дифференциал

Первый дифференциал — это линейная форма, со всеми частными производными и дифференциалами переменных:

\[
$$
df(x) = \sum_{i=1}^n \frac{\partial f}{\partial x_i} dx_i = \sum_{i=1}^n x_i \|x\|_2\, dx_i
$$
\]

В матричной записи (если \(x\) — вектор-столбец, а \(dx\) — вектор-столбец приращений):

\[
$$
df(x) = \|x\|_2\, x^\top dx
$$
\]

# 2.5

Рассмотрим функцию  
$$
f(x) = \|x\|_2, \quad x \in \mathbb{R}^n \setminus \{0\},
$$

где  
$$
\|x\|_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
$$  
— евклидова норма.

---

## Градиент

Градиент — это вектор частных производных:

\[
$$
\nabla f(x) = \left(\frac{\partial f}{\partial x_1}, \dots, \frac{\partial f}{\partial x_n}\right)
$$
\]

Вычислим частную производную по \(x_i\):

\[
$$
\frac{\partial f}{\partial x_i} = \frac{\partial}{\partial x_i} \left(\sum_{k=1}^n x_k^2\right)^{\!1/2}
= \frac{1}{2} \left(\sum_{k=1}^n x_k^2\right)^{\!-1/2} \cdot 2x_i
= \frac{x_i}{\|x\|_2}
$$
\]

**Ответ:**

\[
$$
\boxed{\nabla f(x) = \frac{x}{\|x\|_2}}
$$
\]

---

## Первый дифференциал

Первый дифференциал — это сумма всех частных производных, умноженных на соответствующие дифференциалы:

\[
$$
df(x) = \sum_{i=1}^n \frac{\partial f}{\partial x_i} dx_i = \sum_{i=1}^n \frac{x_i}{\|x\|_2} dx_i
$$
\]

В матричной записи (если \(x\) — вектор-столбец, \(dx\) — вектор приращений):

\[
$$
df(x) = \frac{x^\top dx}{\|x\|_2}
$$
\]

**Ответ:**

\[
$$
\boxed{df(x) = \frac{x^\top dx}{\|x\|_2}}
$$
\]

# 2.6

Рассмотрим функцию

\[
$$
f(x) = \|Ax\|_2, \quad x \in \mathbb{R}^n \setminus \{0\}, \quad A \in \mathbb{R}^{m \times n},
$$
\]

где \($$\|Ax\|_2 = \sqrt{\sum_{i=1}^m (A_i x)^2}$$\) — евклидова норма вектора \(Ax\), а \(A_i\) — \(i\)-я строка матрицы \(A\).

---

## Градиент

Градиент — это вектор частных производных:

\[
$$
\nabla f(x) = \left( \frac{\partial f}{\partial x_1}, \dots, \frac{\partial f}{\partial x_n} \right)
$$
\]

Частная производная по \(x_j\):

\[
$$
\frac{\partial f(x)}{\partial x_j}
= \frac{1}{2\|Ax\|_2} \cdot 2 \sum_{i=1}^m (A_i x) a_{ij}
= \frac{1}{\|Ax\|_2} \sum_{i=1}^m (A_i x) a_{ij}
$$
\]

В матричной форме, сумма \($$\sum_{i=1}^m (A_i x) a_{ij}$$\) — это \(j\)-я компонента вектора \($$A^\top (A x)$$\).

**Итог:**  
\[
$$
\boxed{
\nabla f(x) = \frac{A^\top A x}{\|Ax\|_2}
}
$$
\]

---

## Первый дифференциал

Первый дифференциал — сумма частных производных, умноженных на дифференциалы переменных:

\[
$$
df(x) = \sum_{j=1}^n \frac{\partial f}{\partial x_j} dx_j
= \sum_{j=1}^n \frac{(A^\top A x)_j}{\|Ax\|_2} dx_j
= \frac{(A^\top A x)^\top dx}{\|Ax\|_2}
$$
\]

Также, поскольку \($$A^\top A x = A^\top (A x)$$\), то

\[
$$
(A x)^\top A dx = (A^\top A x)^\top dx
$$
\]

**Итог двумя способами:**  
\[
$$
\boxed{
df(x) = \frac{(A^\top A x)^\top dx}{\|Ax\|_2}
}
$$
\]
или
\[
$$
\boxed{
df(x) = \frac{(A x)^\top A dx}{\|Ax\|_2}
}
$$
\]

# 2.7

Рассмотрим функцию

\[
$$
f(x) = -e^{-x^\top x}, \quad x \in \mathbb{R}^n,
$$
\]

где \($$x^\top x = x_1^2 + x_2^2 + \dots + x_n^2$$\) — скалярное произведение, то же самое, что квадрат евклидовой нормы \($$\|x\|_2^2$$\).

---

## Градиент

Градиент — это вектор частных производных:

\[
$$
\nabla f(x) = \left(\frac{\partial f}{\partial x_1}, \dots, \frac{\partial f}{\partial x_n}\right)
$$
\]

Вычислим частную производную по \(x_j\):

\[
$$
\frac{\partial f(x)}{\partial x_j} = -e^{-x^\top x} \cdot (-2x_j) = 2x_j e^{-x^\top x}
$$
\]

**Ответ:**

\[
$$
\boxed{
\nabla f(x) = 2x\, e^{-x^\top x}
}
$$
\]

---

## Первый дифференциал

Первый дифференциал — это линейная часть разложения \($$f(x + dx) - f(x)$$\):

\[
$$
df(x) = \sum_{j=1}^n \frac{\partial f(x)}{\partial x_j} dx_j = 2e^{-x^\top x} \sum_{j=1}^n x_j dx_j
$$
\]

В матричной записи (если \(x\) — вектор-столбец, а \(dx\) — вектор приращений):

\[
$$
df(x) = 2e^{-x^\top x} x^\top dx
$$
\]

**Ответ:**

\[
$$
\boxed{
df(x) = 2e^{-x^\top x} x^\top dx
}
$$
\]
