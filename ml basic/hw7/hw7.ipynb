# I. Numpy

### Импортируйте NumPy

```python
import numpy as nm
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = nm.zeros(10)
arr[4] = 1
arr_2d = arr.reshape(2, 5)
print(arr_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = nm.arange(10, 50)
arr_reversed = arr[::-1]
even_elements = arr_reversed[arr_reversed % 2 == 0]
print("Развернутый массив:", arr_reversed)
print("Четные элементы:", even_elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr_3x3 = nm.arange(9).reshape(3, 3)
print(arr_3x3)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr_4x3x2 = nm.random.random((4, 3, 2))
print("Массив:\n", arr_4x3x2)
print("Минимум:", arr_4x3x2.min())
print("Максимум:", arr_4x3x2.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = nm.random.random((6, 4))
arr2 = nm.random.random((4, 3))
result = nm.matmul(arr1, arr2)
print("Первый массив (6x4):\n", arr1)
print("Второй массив (4x3):\n", arr2)
print("Результат умножения (6x3):\n", result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr_7x7 = nm.random.random((7, 7))
mean = arr_7x7.mean()
std = arr_7x7.std()
arr_norm = (arr_7x7 - arr_7x7.min()) / (arr_7x7.max() - arr_7x7.min())
print("Среднее:", mean)
print("Стандартное отклонение:", std)
print("Нормализованный массив (от 0 до 1):\n", arr_norm)
```

### Найдите максимальное значение 'total_bill'

```python
print("Максимальное значение 'total_bill':", tips['total_bill'].max())
```

### Найдите количество курящих людей

```python
print("Количество курящих людей:", tips[tips['smoker'] == 'Yes'].shape[0])
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print(tips.groupby('day')['total_bill'].mean())
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
filtered = tips[tips['total_bill'] > tips['total_bill'].median()]
print(filtered.groupby('sex')['tip'].mean())
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker_bin'] = tips['smoker'].map({'No': 0, 'Yes': 1})
print(tips[['smoker', 'smoker_bin']].tail())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], bins=10, kde=True, color='green')
plt.title("Гистограмма распределения 'total_bill'")
plt.xlabel('Total Bill')
plt.ylabel('Количество')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Scatterplot: total_bill vs tip")
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Взаимосвязь между 'total_bill' и 'day'")
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=tips, x='tip', hue='time', bins=5, kde=True, element='step', stat='density', common_norm=False)
plt.title("Гистограммы распределения 'tip' для Lunch и Dinner")
plt.xlabel('Tip')
plt.ylabel('Плотность')
plt.legend(title='Time')
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.scatterplot(
    data=tips[tips['sex'] == 'Male'],
    x='total_bill', y='tip', hue='smoker',
    palette='Set1', ax=axes[0]
)
axes[0].set_title("Male: total_bill vs tip")
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

sns.scatterplot(
    data=tips[tips['sex'] == 'Female'],
    x='total_bill', y='tip', hue='smoker',
    palette='Set1', ax=axes[1]
)
axes[1].set_title("Female: total_bill vs tip")
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```
